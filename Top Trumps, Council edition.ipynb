# Population Data

From: http://stat.data.abs.gov.au/

```python
import pandas as pd
filename = "data/populations.csv"
```

```python
populations = pd.read_csv(open(filename, encoding='utf8', errors='ignore'), skiprows=5)
```

```python
def get_council_name(mixed):
    tokens = mixed.split(" (")
    return tokens[0].strip()
    
    
populations = populations.drop(labels=0).drop("Unnamed: 1", axis=1)
populations["Time"] = populations["Time"].apply(get_council_name)
```

```python
populations.set_index("Time", inplace=True)
populations.index.name = "Region"
```

```python
populations.sort_index(inplace=True)
```
